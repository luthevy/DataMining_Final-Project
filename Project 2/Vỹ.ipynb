```python
style = ttk.Style()

instruction=Label(fram2_1,text="CHỌN MỘT THUỘC TÍNH",font='Calibri 16 bold')
instruction.pack(side="top")

#Tạo data list là các thuộc tính của dataset (dùng OptionMenu)
options=df.columns.values
clicked=StringVar()
clicked.set(options[0])
drop=OptionMenu(fram2_1,clicked,*options)
drop.pack(pady=20)
drop.config(width=20)

instruction=Label(fram2_2,text="TRỰC QUAN HOÁ",font='Calibri 16 bold')
instruction.pack(side="top")

# Biến tmp dùng để làm giá trị khởi đầu tạm thời (tránh lỗi)
tmp="texture_mean"

# chia các cột theo loại: numeric và categorical
categorical_cols = ['diagnosis']
numeric_cols = ['radius_mean', 'texture_mean', 'perimeter_mean', 'area_mean',
                    'smoothness_mean', 'compactness_mean', 'concavity_mean',
                    'concave points_mean', 'symmetry_mean', 'fractal_dimension_mean',
                    'radius_se', 'texture_se', 'perimeter_se', 'area_se', 'smoothness_se',
                    'compactness_se', 'concavity_se', 'concave points_se', 'symmetry_se',
                    'fractal_dimension_se', 'radius_worst', 'texture_worst',
                    'perimeter_worst', 'area_worst', 'smoothness_worst',
                    'compactness_worst', 'concavity_worst', 'concave points_worst',
                    'symmetry_worst', 'fractal_dimension_worst']

fig1 = plt.Figure(figsize=(1,1), dpi=100)
ax1 = fig1.add_subplot(111)
bar = FigureCanvasTkAgg(fig1, fram2_2)
bar.get_tk_widget().pack(side=tk.LEFT, fill=tk.BOTH)
fig2 = plt.Figure(figsize=(1,1), dpi=100)
ax2 = fig2.add_subplot(111)
pie = FigureCanvasTkAgg(fig2, fram2_2)
pie.get_tk_widget().pack(side=tk.LEFT, fill=tk.BOTH)
fig3 = plt.Figure(figsize=(1,1), dpi=100)
ax3 = fig3.add_subplot(111)
line = FigureCanvasTkAgg(fig3, fram2_2)
line.get_tk_widget().pack(side=tk.LEFT, fill=tk.BOTH)
fig4 = plt.Figure(figsize=(1,1), dpi=100)
ax4 = fig4.add_subplot(111)
box = FigureCanvasTkAgg(fig4, fram2_2)
box.get_tk_widget().pack(side=tk.LEFT, fill=tk.BOTH)
fig5 = plt.Figure(figsize=(1,1), dpi=100)
ax5 = fig5.add_subplot(111)
hist = FigureCanvasTkAgg(fig5, fram2_2)
hist.get_tk_widget().pack(side=tk.LEFT, fill=tk.BOTH)
fig6 = plt.Figure(figsize=(1,1), dpi=100)
ax6 = fig6.add_subplot(111)
violin = FigureCanvasTkAgg(fig6, fram2_2)
violin.get_tk_widget().pack(side=tk.LEFT, fill=tk.BOTH)
fig7 = plt.Figure(figsize=(1,1), dpi=100)
ax7 = fig7.add_subplot(111)
ogive = FigureCanvasTkAgg(fig7, fram2_2)
ogive.get_tk_widget().pack(side=tk.LEFT, fill=tk.BOTH)

#Thực thi lệnh cập nhật dữ liệu cho dòng
def my_show():
    tmp = clicked.get()
    
    if tmp in categorical_cols:
        # Bar Chart
#             fig1 = plt.Figure(figsize=(5,5), dpi=100)
#             ax1 = fig1.add_subplot(111)
#             bar = FigureCanvasTkAgg(fig1, fram2_2)
#             bar.get_tk_widget().pack(side=tk.LEFT, fill=tk.BOTH)
        sns.countplot(x=df[tmp],ax=ax1)
        ax1.set_xlabel(tmp)
        ax1.set_title('Bar Chart')

        # Pie Chart
#             fig2 = plt.Figure(figsize=(5,5), dpi=100)
#             ax2 = fig2.add_subplot(111)
#             pie = FigureCanvasTkAgg(fig2, fram2_2)
#             pie.get_tk_widget().pack(side=tk.LEFT, fill=tk.BOTH)
        value_counts = df[tmp].value_counts()
        ax2.pie(value_counts,autopct='%d',labels=value_counts.index)
        ax2.set_title('Pie Chart')

    elif tmp in numeric_cols:
        # Line Chart
#             fig1 = plt.Figure(figsize=(5,5), dpi=100)
#             ax1 = fig1.add_subplot(111)
#             line = FigureCanvasTkAgg(fig1, fram2_2)
#             line.get_tk_widget().pack(side=tk.LEFT, fill=tk.BOTH)
        df[tmp].plot(ax=ax1)
        ax1.set_xlabel(tmp)
        ax1.set_ylabel('Density')
        ax1.set_title('Line Chart')

        # Box Plot
#             fig2 = plt.Figure(figsize=(5,5), dpi=100)
#             ax2 = fig1.add_subplot(111)
#             box = FigureCanvasTkAgg(fig2, fram2_2)
#             box.get_tk_widget().pack(side=tk.LEFT, fill=tk.BOTH)
        sns.boxplot(x=df[tmp],ax=ax2)
        ax2.set_xlabel(tmp)
        ax2.set_title('Box Plot')

        # Histogram
#             fig3 = plt.Figure(figsize=(5,5), dpi=100)
#             ax3 = fig3.add_subplot(111)
#             hist = FigureCanvasTkAgg(fig3, fram2_2)
#             hist.get_tk_widget().pack(side=tk.LEFT, fill=tk.BOTH)
        sns.histplot(x=df[tmp],kde=True,ax=ax3)
        ax3.set_xlabel(tmp)
        ax3.set_ylabel('Density')
        ax3.set_title('Histogram')

        # Violin Plot
#             fig4 = plt.Figure(figsize=(5,5), dpi=100)
#             ax4 = fig1.add_subplot(111)
#             violin = FigureCanvasTkAgg(fig4, fram2_2)
#             violin.get_tk_widget().pack(side=tk.LEFT, fill=tk.BOTH)
        sns.violinplot(x=df[tmp],ax=ax4)
        ax4.set_xlabel(tmp)
        ax4.set_title('Violin Plot')

        # Ogive Graph
#             fig5 = plt.Figure(figsize=(5,5), dpi=100)
#             ax5 = fig5.add_subplot(111)
#             ogive = FigureCanvasTkAgg(fig5, fram2_2)
#             ogive.get_tk_widget().pack(side=tk.LEFT, fill=tk.BOTH)
        # calculating frequency and class interval
        values, base = np.histogram(df[tmp], bins=np.linspace(0,df[tmp].max(),10))
        # calculating cumulative sum
        cumsum = np.cumsum(values)
        # plotting  the ogive graph
        sns.lineplot(x=base[1:], y=cumsum, color='red', marker='o', linestyle='-',ax=ax5)
        # formatting
        ax5.set_title('Ogive Graph')
        ax5.set_xlabel(tmp)
        ax5.set_ylabel('Cumulative Frequency')
        
def clear():
    bar.get_tk_widget().delete(0, END)
    pie.get_tk_widget().delete(0, END)
    line.get_tk_widget().delete(0, END)
    box.get_tk_widget().delete(0, END)
    hist.get_tk_widget().delete(0, END)
    violin.get_tk_widget().delete(0, END)
    ogive.get_tk_widget().delete(0, END)

l2 = Button(fram2_1,  text="Xác nhận", width=30,command=my_show,fg='YELLOW',bg='GREEN', font='Calibri 12 bold') 
l3 = Button(fram2_1,  text="Clear", width=30,command=my_show,fg='YELLOW',bg='GREEN', font='Calibri 12 bold')  

#Thực thi lệnh widget
l2.pack(pady=30)
l3.pack(pady=20)
# stats_result.pack(pady=20)
```
